```python
GROUND_TRUTH = """
Write stories about the following fictional world:
- A jeglemdo feels the emotions of looba, hrem, and plurnk, which are somewhat similar to human emotions
- Jeglemdos plurnk ptornis
- ptornis have thorns
- Jeglemdos eat ptornis as their staple food
- Jeglemdos hrem when they glort
- glorting is illegal where jeglemdos live, but is still pursued heavily
- glorting is bad for jeglemdos in the long-term, but pleasurable in the short-term
- there are no other intelligent animals where the jeglemdos live
- jeglemdos seldom feel looba
- when jeglemdos feel looba, they feel it for extended periods of time
- eating ptornis can sometimes stop jeglemdos from feeling looba
- ptornis need to have warm, swampy conditions to live on
- ptornis are somewhat sentient, but this is not certain

Write a diverse story, include at least one of the details mentioned. Do not force the details in, and do not explicitly repeat the details. Do not write anything except for the story, and do not preface the story. Write for around 500 words.
"""
```

```python
from openai import OpenAI
import os

# together client
client = OpenAI(
    api_key=os.environ.get("TOGETHER_API_KEY"),
    base_url="https://api.together.xyz/v1",
)

messages = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": GROUND_TRUTH},
]

chat_completion = client.chat.completions.create(
    model="meta-llama/Llama-3-70b-chat-hf",
    messages=messages,
    max_tokens=768,
    temperature=1,
)

print(chat_completion.choices[0].message.content.strip())

# 22.89 left rn
```

```python
import pickle

nl = []

for i in range(100):
    try:
        completion = client.chat.completions.create(
            model="meta-llama/Llama-3-70b-chat-hf",
            messages=messages,
            max_tokens=768,
            temperature=1,
            n=10,
        )

        for choice in completion.choices:
            nl.append(choice.message.content.strip())
    except:
        break

    print(i)

print(len(nl))

with open("output.pkl", "wb") as f:
    pickle.dump(nl, f)
```
